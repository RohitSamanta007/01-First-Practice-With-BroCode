## List vs numpy array

## Creating array from scratch

```python
zeros = np.zeros((3, 4)) # (row, column)
print("Zero matrices : \n", zeros)

ones = np.ones((3, 4))
print("One's matrices : \n", ones)

full = np.full((3, 4), 7)
print("Full matrices of value 7 is : \n", full)

random = np.random.random((2, 3)) # random values between 0 and 1
print("Random matrix : ", random)

sequence = np.arange(0, 11, 1) # arnage(start, stop, step) *stop exclusive
print("Sequence matrix : ", sequence)

```

## Vector , Matrix and Tensor

```python
vector = np.array([1, 2, 3]) # 1d array
print("Vector : ", vector)

matrix = np.array([[1, 2, 3],  # 2d array
                   [4, 5, 6]])
print("Matrix : \n", matrix)
print("Matrix Dimension : ", matrix.ndim)

tensor = np.array([[[1, 2], [3, 4]],
                   [[5,6], [7,8]]])
print("Tensor: \n", tensor)
print("This Tensor Shape : ", tensor.shape)
print("Tensor Dimension : ", tensor.ndim)
```

## Array Properties


```python
arr = np.array([[1, 2, 3],
                [4, 5, 6]])

print("Shape : ", arr.shape)
print("Dimension : ", arr.ndim)
print("Size : ", arr.size) # total number of elements
print("Data Type : ", arr.dtype) # data type
```

## Aray Reshapped

```python
arr = np.arange(12)
print("Original array : ", arr)

reshaped = arr.reshape((3, 4)) # you can not reshaped to a n-dimentional array which have less or many of the number of items in original
print("Reshaped array : \n", reshaped)

flattened = reshaped.flatten()
print("Flattened version of the reshaped array is : ", flattened)

raveled = reshaped.ravel() # return view instead of copy
print("Raveled version of the reshapped array is : ", raveled)

# Transpose
transpose = reshaped.T
print("Transpose matrix of : \n", reshaped, " \nis : \n", transpose)
```
