### Advance Operation with Business Examples

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Chai Point
])

print("Salas data shape : ", sales_data.shape)
print("Sample data for 1st 3 resturent : \n", sales_data[0:3]) # [start: end]
print("Sample data for 1st 3 resturent : \n", sales_data[:, 1:]) # here we access data by row and column [start, end]-> all rows and start from the second column to last column for all rows
```

```python
# total sale per year 
print(np.sum(sales_data, axis=0)) # do column wise addtion and get output as 1D array
print(np.sum(sales_data, axis=1)) # do row wise addtion and get output as 1D array

year_total = np.sum(sales_data[:, 1:], axis=0)
print("Year total : ", year_total)

# minimum
min_sales = np.min(sales_data[:, 1:], axis=1)
print("Mininum sales per resturent : ", min_sales)

# maximum
max_sales = np.max(sales_data[:, 1:], axis=0)
print("Maximum sale of resturent per years : ", max_sales)

# Average 
avg_sales = np.mean(sales_data[:, 1:], axis=1)
print("Average sales per resturent : ", avg_sales)

# cumulative sum 
cumsum = np.cumsum(sales_data[:, 1:], axis=1)
print("The cumulative sum of sales per resturent : \n", cumsum)

#plot sale data
plt.figure(figsize=(8, 6))
plt.plot(np.mean(cumsum, axis=0))
plt.title("Average cumulative sales accross all resturent : ")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```

## Vector Operations


```python
vector1 = np.array([1, 2, 3, 4, 5])
vector2 = np.array([6, 7, 8, 9, 10])

print("Vector addition : ", vector1 + vector2)
print("Vector multiplication : ", vector1 * vector2)
print("Dot product : ", np.dot(vector1, vector2))

angle = np.arccos(np.dot(vector1, vector2) / (np.linalg.norm(vector1) * np.linalg.norm(vector2))) # Rule : sin(a,b) = (a.b)/(||a|| . ||b||)
print("The angle between two vector is :", angle)
```

## Vectorized operations
